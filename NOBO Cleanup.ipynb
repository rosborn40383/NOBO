```python
import pandas as pd #Great for formatting and reimaging data
import numpy as np #Great for statistical functions
#Just a general cleanup of the data that should make things easier going forward. 
#Going to use this csv in another notebook to test some other things
```

```python
df = pd.read_csv('NOBO_INFO_NEW.csv') #Read data into Pandas
df_copy = df.copy() #Making a copy incase
```

```python
df.head() #Quick header to see they data I am working with.
```

```python
columns_drop = ["SORT:(D)ESCENDING (Z)IP CODE", "NUMBER OF NOBOS","7TH LINE OF NAME & ADDRESS" ] #Taking out columns
df = df.drop(columns=columns_drop) # Drop these columns since they provide no relevant info
```

```python
df.head() #Checking to see if it worked
```

```python
column_count = '6TH LINE OF NAME & ADDRESS' #Getting a count of the 6th line of name column
count_result = df[column_count].value_counts() #want a count because it's very empty
print(count_result)
```

```python
df = df.rename(columns={'1ST LINE OF NAME & ADDRESS': 'CLIENT'}) #Renaming for accuracy and easier to work with
df.head() #Always check my work
```

```python
df['Inv_type'] = 'Other'  # Default value, if it doesn't meet criteria then will be other
# Define rules to categorize based on keywords, had to look through some of the file
df.loc[(df['CLIENT'].str.contains('FUND|COUNTY|CITY', case=False)) | (df['2ND LINE OF NAME & ADDRESS'].str.contains('FUND|COUNTY|CITY', case=False)), 'Inv_type'] = 'Bond Fund'
df.loc[(df['CLIENT'].str.contains('INSURANCE|ASSURANCE|KSKJ|HEALTH|NATIONWIDE', case=False)) | (df['2ND LINE OF NAME & ADDRESS'].str.contains('INSURANCE|ASSURANCE|KSKJ|HEALTH|NATIONWIDE', case=False)), 'Inv_type'] = 'Insurance Company'
df.loc[(df['CLIENT'].str.contains('INVESTMENT|INVESTING|CAPITAL|TRUST|WEALTH|SECURITIES', case=False)) | (df['2ND LINE OF NAME & ADDRESS'].str.contains('INVESTMENT|INVESTING|CAPITAL|TRUST|WEALTH|SECURITIES', case=False)), 'Inv_type'] = 'Investment Firm'
df.loc[(df['CLIENT'].str.contains('RETAIL|BANK|IRA|SCHWAB', case=False)) | (df['2ND LINE OF NAME & ADDRESS'].str.contains('RETAIL|BANK|IRA|SCHWAB', case=False)), 'Inv_type'] = 'Retail'
# check work, this is probably where it could get classification wrong because it searches based on the keywords
print(df[['CLIENT', '2ND LINE OF NAME & ADDRESS', 'Inv_type']])
```

```python
df.head()
```

```python
#Goal of this is to remove '6th line of name & address' and combine the data into the previous column, less columns is better
df.loc[:, 'COMBINED ADDRESS'] = df['5TH LINE OF NAME & ADDRESS'].astype(str) + ' ' + df['6TH LINE OF NAME & ADDRESS'].astype(str)
#drop any missing data values, loc is important because it keeps the data in the same shape
df = df.dropna(subset=['6TH LINE OF NAME & ADDRESS'])
```

```python
df #check the dataframe
```

```python
df.loc[:, 'COMBINED ADDRESS'] = df['4TH LINE OF NAME & ADDRESS'].astype(str) + ' ' + df['COMBINED ADDRESS'].astype(str)
#Trying to do the same with the previous column
```

```python
df
```

```python
columns_to_drop = ['4TH LINE OF NAME & ADDRESS', '5TH LINE OF NAME & ADDRESS', '6TH LINE OF NAME & ADDRESS'] #bye columns
df = df.drop(columns=columns_to_drop)
```

```python
# Find the index of the 'Inv_type' column
inv_type_index = df.columns.get_loc('Inv_type') #Want the new column to be moved 
new_position = 8
df.insert(new_position, 'Inv_type', df.pop('Inv_type')) # Move the 'Inv_type' column to the new position
```

```python
df
```

```python
columns_to_drops = ['CUSIP.1'] #bye columns, this was a duplicate column
df = df.drop(columns=columns_to_drops) #should have removed them earlier but got error since itn't there anymore
df
```

```python
#try to add some of the address data from "3rd" to the combined address without adding non-address data
condition = df['3RD LINE OF NAME & ADDRESS'].str.strip().str.contains('\d', na=False)
df['COMBINED ADDRESS'] = np.where(condition, df['COMBINED ADDRESS'].astype(str) + ' ' + df['3RD LINE OF NAME & ADDRESS'].fillna('').astype(str), df['COMBINED ADDRESS'])
```

```python
df
```

```python
df = df.rename(columns={
    'COMBINED ADDRESS': 'ADDRESS',
    'ZIP CODE FOR SORTING': 'ZIP CODE',
    'NUMBER OF SHARES': 'TOTAL SHARES',
    'ISSUER NAME': 'ISSUER',
    'Inv_type': 'CLASSIFICATION',
    'RECORD DATE': 'DATE',
    '2ND LINE OF NAME & ADDRESS': 'CONTACT/ADDRESS',
    '3RD LINE OF NAME & ADDRESS': 'FULL ADDRESS'
})
#rename columns since it is easier to work with and more practical
```

```python
df
```

```python
df['ADDRESS'] = df['FULL ADDRESS'].fillna('') + ' ' + df['ADDRESS'].fillna('') #change full address to address
df.insert(7, 'ADDRESS', df.pop('ADDRESS')) #Move it into a different position
df.drop('FULL ADDRESS', axis=1, inplace=True) #Drop the Full address column since it was replaced by address column
```

```python
df.drop('DATE', axis=1, inplace=True) #Also dropped date column since it didn't have useable info, all the same (111323)
df
```

```python
df.drop('ISSUER', axis=1, inplace=True) #Dropped issuer since all the same and will allow it to proccess quicker
df
```

```python
# changing around the columns for swag
df = df[['JOB NUMBER','CUSIP', 'CLIENT', 'CLASSIFICATION','SHARES', 'TOTAL SHARES' ,'CONTACT/ADDRESS', 'ADDRESS', 'ZIP CODE', '# OF NOBOS']]
df
```
